### Doodle Classifier Project

### Dependencies

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Loading the data

```python
train_df = tf.keras.preprocessing.image_dataset_from_directory("Data/train",color_mode="grayscale")

class_names = train_df.class_names
print(class_names)
```

```python
for x in range(3):
    batch = train_df.as_numpy_iterator().next()
    fig,ax = plt.subplots(ncols = 10,figsize = (20,20))
    for idx,img in enumerate(batch[0][:10]):
    	ax[idx].imshow(img.astype(int),cmap = "gray")
    	ax[idx].title.set_text(class_names[batch[1][idx]])
```
